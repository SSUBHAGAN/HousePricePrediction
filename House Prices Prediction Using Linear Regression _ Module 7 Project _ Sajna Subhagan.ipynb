## 1. Introduction

The objective is to analyze the existing house sale prices data and to **build a model for predicting the house sale prices.**<br>

Steps Involved:<br>
Importing the dataset followed by exploratory data analysis to find insights into the data.The dataset had 100 rows and 82 columns.<br>
**Check for duplicates, nulls** and analyze the data to find patterns and correlations between the dependent & independent variables.<br>
Data Preparation involves treating the nulls, converting the categorical columns to numerical or one hot encoding.<br>

The algorithm used for predicting the house prices is **Linear Regression.** <br>
Linear Regression is a **supervised machine learning algorithm** which performs a regression task. Regression models a target prediction value based on independent variables. 6 Models were developed with various combinations of the independent variables. The independent variables or features were chosen based on their strong correlation with the target variable - Sale price. **The independent variables that exhibited strong correlation to one another were eliminated to avoid multicolliearity**.Multicollinearity is the occurrence of high intercorrelations among two or more independent variables in a multiple regression model. Multicollinearity can lead to skewed or misleading results when trying to determine how well each independent variable can be used most effectively to predict or understand the dependent variable in a statistical model. <br>
<br>
The best model was chosen based on the **R- squared scores** which determines the proportion of variance in the dependent variable that can be explained by the independent variable.The results were supported by **Root mean Square error scores**.The Rsquared scores were **0.89 for the training dataset** and **0.86 for the testing dataset** and uses 17 features/independent variables and had the least **RMSE** scores.

### **3.1 Linear Regression**


Linear regression is a supervised machine learning algorithm that is used to predict values within a continuous range (i.e., stocks’ price and height) rather than classifying them into categories (i.e., male or female).
The values that you want to predict are the dependent variable(s). The predictions can be made using one or more independent variables.

Linear regression is a linear approach for modelling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables). The case of one explanatory variable is called simple linear regression; for more than one, the process is called multiple linear regression.This term is distinct from multivariate linear regression, where multiple correlated dependent variables are predicted, rather than a single scalar variable.

**Mathematically the relationship can be represented with the help of following equation −**

>#### Y = mX + b<br>
>*Y is the dependent variable we are trying to predict.*<br>
>*X is the dependent variable we are using to make predictions.*<br>
>*m is the slop of the regression line which represents the effect X has on Y*<br>
>*b is a constant, known as the Y-intercept. If X = 0,Y would be equal to b.*

A linear relationship will be called positive if both independent and dependent variable increases. 

A linear relationship will be called positive if independent increases and dependent variable decreases.
<br>
<br>
Linear regression is of the following two types −
<br>
**Simple Linear Regression -**
which predicts a response using a single feature. The assumption in SLR is that the two variables are linearly related.

Python implementation
We can implement linear regression in Python by using importing liner_model from sklearn library.

Python implementation example

First, we will start with importing necessary packages as follows −
Next, we will load the dataset.
Split the dataset into training & testing datset - X-train, X-test
Next, we need to split the target into training and testing sets as follows - Y-train,Y-test 
Now, to train the model we need to create linear regression object as follows −

lr = linear_model.LinearRegression()

Next, train the model using the training sets as follows −

lr.fit(X_train, y_train)

Next, make predictions using the testing set as follows −

y_pred = regr.predict(X_test)<br>
<br>
<br>
**Multiple linear regression** can be viewed as a natural extension of simple linear regression. The main difference is that now multiple sets of data are used to make the prediction.

Therefore, the equation for multiple linear regression becomes:

>#### Y = m1X1 +m2X2 +m3x3.....+mnxn+ b<br>
>*where x1, x2....xn  represent the data points of all the independent variables used to make the prediction,m1,m2....,mn  are the weights, and b is the bias.*<br>


### **Assumptions**

The following are some assumptions about dataset that is made by Linear Regression model −

- *Multi-collinearity* − Linear regression model assumes that there is very little or no multi-collinearity in the data. Basically, multi-collinearity occurs when the independent variables or features have dependency in them.

- *Auto-correlation* − Another assumption Linear regression model assumes is that there is very little or no auto-correlation in the data. Basically, auto-correlation occurs when there is dependency between residual errors.

- *Relationship between variables* − Linear regression model assumes that the relationship between response and feature variables must be linear.
